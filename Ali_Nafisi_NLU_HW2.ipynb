## Load data

```python
import pandas as pd

splits = {'train': 'plain_text/train-00000-of-00001.parquet', 'test': 'plain_text/test-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/fancyzhx/yelp_polarity/" + splits["train"])
pd.set_option('display.max_colwidth', None)
df.head()
```

## Part 1

### List the dates in the sentences

```python
import re
import pandas as pd

date_pattern = r'(\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b|\b(?:January|February|March|April|May|June|July|August|September|October|November|December)\s+\d{1,2},\s+\d{4}\b)'

def extract_dates(text):
    return re.findall(date_pattern, text)

df['dates'] = df['text'].apply(extract_dates)

for i in range(1000):
    if df.iloc[i]['dates'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['dates'])
        print('-'*100)
```

### List the hours in the sentences

```python
time_pattern = r'\b\d{1,2}:\d{2}\s?(?:AM|PM|am|pm)?\b'

def extract_times(text):
    return re.findall(time_pattern, text)

df['times'] = df['text'].apply(extract_times)

for i in range(1000):
    if df.iloc[i]['times'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['times'])
        print('-'*100)
```

### List the web links in the sentences

```python
url_pattern = r'(https?://\S+|www\.\S+)'

def extract_urls(text):
    return re.findall(url_pattern, text)

df['urls'] = df['text'].apply(extract_urls)

for i in range(1000):
    if df.iloc[i]['urls'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['urls'])
        print('-'*100)
```

### List names starting with "Dr".

```python
dr_pattern = r'\bDr\.\s+[A-Z][a-z]*\b'

def extract_dr_names(text):
    return re.findall(dr_pattern, text)

df['dr_names'] = df['text'].apply(extract_dr_names)

for i in range(1000):
    if df.iloc[i]['dr_names'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['dr_names'])
        print('-'*100)
```

### List dollar price values

```python
price_pattern = r'\$\d+(?:\.\d{1,2})?'

def extract_prices(text):
    return re.findall(price_pattern, text)

df['prices'] = df['text'].apply(extract_prices)

for i in range(1000):
    if df.iloc[i]['prices'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['prices'])
        print('-'*100)
```

## Part 2

### Step 1: Remove web links

```python
def remove_urls(text):
    return re.sub(url_pattern, '', text)

df['clean_text'] = df['text'].apply(remove_urls)

for i in range(1000):
    if df.iloc[i]['urls'] != []:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['clean_text'])
        print('-'*100)
```

### Step 2: Remove emojis

```python
emoji_pattern = re.compile(
    "["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
    u"\U00002500-\U00002BEF"  # chinese char
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    u"\U0001f926-\U0001f937"
    u"\U00010000-\U0010ffff"
    u"\u2640-\u2642"
    u"\u2600-\u2B55"
    u"\u200d"
    u"\u23cf"
    u"\u23e9"
    u"\u231a"
    u"\ufe0f"  # dingbats
    u"\u3030"
    "]+",
    re.UNICODE
)

def remove_emojis(text):
    return emoji_pattern.sub(r'', text)

df['clean_text'] = df['clean_text'].apply(remove_emojis)


# There was no emjies in the text
for i in range(len(df)):
    if re.findall(emoji_pattern, df.iloc[i]['text']):
        print(df.iloc[i]['text'])
        print(df.iloc[i]['clean_text'])
        print('-'*100)
```

### Step 3: Remove numbers

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

df['clean_text'] = df['clean_text'].apply(remove_numbers)

for i in range(1000):
    if bool(re.search(r'\d', df.iloc[i]['text'])):
        print(df.iloc[i]['text'])
        print(df.iloc[i]['clean_text'])
        print('-'*100)
```

### Step 4: Remove stopwords

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

df['clean_text'] = df['clean_text'].apply(remove_stopwords)

for i in range(1000):
    if [word for word in word_tokenize(df.iloc[i]['text']) if word.lower() not in stop_words]:
        print(df.iloc[i]['text'])
        print(df.iloc[i]['clean_text'])
        print('-'*100)
```

### Step 5: Remove punctuation

```python
import string

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

df['clean_text'] = df['clean_text'].apply(remove_punctuation)

for i in range(100):
    print(df.iloc[i]['text'])
    print(df.iloc[i]['clean_text'])
    print('-'*100)
```

### Step 6: Lemmatization

```python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = word_tokenize(text)
    lemmatized = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(lemmatized)

df['clean_text'] = df['clean_text'].apply(lemmatize_text)

for i in range(100):
    print(df.iloc[i]['text'])
    print(df.iloc[i]['clean_text'])
    print('-'*100)
```

### Step 7: Count the total words and find the 10 most repeated words

```python
from collections import Counter

all_text = ' '.join(df['clean_text'])

all_words = word_tokenize(all_text.lower())

word_counts = Counter(all_words)

total_words = sum(word_counts.values())
print(f"Number of all words: {total_words}")

top_10 = word_counts.most_common(10)
print("10 most common words:")
for word, count in top_10:
    print(f"{word}: {count}")
```
